## Create `income_dict`, a dictionary that uses the numerical codes in the income bucket column as keys and the matching descriptions ('Total', 'Under $1', 'Between 1 and $10,000', 
 'Between 10,000 and $25,000','Between 25,000 and $50,000', 'Between 50,000 and $75,000', 'Between 75,000 and $100,000', 'Between 100,000 and $200,000', '$200,000 or more') as values.

```python
income_dict = {1: 'Under $1', 2: 'Between 1 and $10,000', 3: 'Between 10,000 and $25,000', 4: 'Between 25,000 and $50,000', 5: 'Between 50,000 and $75,000', 6: 'Between 75,000 and $100,000', 7:'Between 100,000 and $200,000', 8: '$200,000 or more'}
```

```python
income_dict
```

 ## Pass the dictionary as an argument to the pandas `replace()` method to change the `income_bucket` column so that it uses descriptive text instead of the numeric code.

```python
income_tennessee_renamed = income_tennessee_cut.replace({"income_bucket": income_dict})
```

 ### Make sure my replace() command worked

```python
income_tennessee_renamed
```

 ## Use the pandas `groupby()` method to group the data by county and get the `sum()` of all numeric columns for that county.

```python
counties_sum = income_tennessee_renamed.groupby(['county']).sum()
```

 ## Be sure to also `reset_index()` so that our aggregated data is re-indexed to begin at 0. Save this as a DataFrame called `income_county_agg`

```python
income_county_agg = counties_sum.reset_index()
```

 ## Look at the first few rows.

```python
income_county_agg.head()
```

 # Begin Week 3

 ## Write a for-loop to iterate through the `income_county_agg` DataFrame and create a list of average incomes for each county. Assign this list to a new column in `income_county_agg`
 ### First create an empty list called `avg_inc_list`

```python
avg_inc_list = []
```

```python
#Begin the for-loop to iterate through `income_county_agg`
for index, row_values in income_county_agg.iterrows():
    #Store the row's total income in a variable called `income`. Remember to multiply total income by 1000 before storing it in order to get the actual amount.
    income = row_values.total_inc_amt * 1000
    #Store each row's `return_count` in a variable called `count`
    count = row_values.returns_with_total_inc
    #divide `income` by `count`, round it to the nearest whole number, and `append` it to `avg_inc_list`.
    avg_inc_list.append(round(income/count))
```

 ## Assign `avg_income_list` to a new column in `income_county_agg` called `avg_income`.

```python
#note, if you dont reset the empty list (avg_inc_list), the indexes won't match up
income_county_agg['avg_income'] = pd.Series(avg_inc_list, index=income_county_agg.index)
```
